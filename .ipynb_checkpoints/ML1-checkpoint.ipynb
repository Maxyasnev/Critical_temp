```python
import pandas as pd
football = pd.read_csv('football.csv')
```

```python
football[football.Wage > football.Wage.mean()].SprintSpeed.mean() #В датасете о футболистах (football.csv) найди среднюю скорость (SprintSpeed) футболистов, зарплата (Wage) которых выше среднего.  Ответ округли до сотых.
```

```python
football[football.Wage < football.Wage.mean()].SprintSpeed.mean() #В датасете о футболистах ( football.csv  ) найди среднюю скорость (SprintSpeed) футболистов, зарплата (Wage) которых ниже среднего.  Ответ округли до сотых.
```

```python
football[(football.Composure > (football.Composure.max()*0.9)) & (football.Reactions > (football.Reactions.max()*0.9))].Age.min() #Укажи возраст (Age) самого молодого игрока из датасета о футболистах (football.csv), у которого хладнокровие (Composure) и реакция (Reactions) превышают 90% от максимального значения.
```

```python
football[football.Value > football.Value.mean()].Nationality.describe().top #Из какой страны (Nationality) больше всего игроков в датасете о футболистах (football.csv), чья стоимость (Value) превышает среднее значение?
```

```python
age = football.Age.mean() #запсиать в перменное среднее значение возраста футболистов и округлить до целых
print(round(age))
```

```python
import pandas as pd #HOMEWORK
football = pd.read_csv('football.csv')
n = football.Nationality.max()
age = football.Age.mean()
print(round(age), n)
```

```python
n = football[football.Wage > football.Wage.mean()].Nationality.max()
speedH = football[football.Wage > football.Wage.mean()].SprintSpeed.mean()
speedL = football[football.Wage < football.Wage.mean()].SprintSpeed.mean()
age = football[(football.Composure > (football.Composure.max()*0.9)) & (football.Reactions > (football.Reactions.max()*0.9))].Age.min() 
print(round(speedH), round(speedL), round(age), n)
```
